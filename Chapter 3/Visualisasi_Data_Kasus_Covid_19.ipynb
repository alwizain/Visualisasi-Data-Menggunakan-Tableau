# Statistik Deskriptif

# Data assessing

```python
# menilai data
data_covid.info()
data_covid.isna().sum()
print("Jumlah duplikasi: ", data_covid.duplicated().sum())
```

# Data Cleaning

```python
# Menghapus Variabel yang tidak diperlukan
data_covid=data_covid.drop(['City or Regency','Time Zone','Special Status'], axis=1)
data_covid.head()
```

```python
# mengatasi missing value
data_covid=data_covid.rename(columns={'Total Cities':'Total_Cities','Total Urban Villages':'Total_Urban', 'Total Rural Villages':'Total_Rural'})
# kolom Province
data_covid.Province.fillna(value="Indonesia", inplace=True)
# kolom Island
data_covid.Island.fillna(value="Indonesia", inplace=True)
# Kolom Total Cities
data_covid.Total_Cities.fillna(value=data_covid.Total_Cities.median(), inplace=True)
# Kolom Total Urban Villages
data_covid.Total_Urban.fillna(value=data_covid.Total_Urban.median(), inplace=True)
# Kolom Total Rural Villages
data_covid.Total_Rural.fillna(value=data_covid.Total_Rural.median(), inplace=True)
```

```python
# mengatasi kesalahan tipe data
# Kolom Date
datetime_columns=["Date"]
for column in datetime_columns :
  data_covid[column]=pd.to_datetime(data_covid[column])
data_covid.info()
```

# Data Processing

```python
# data covid
data_covid = data_covid.loc[data_covid['Province'] != 'Indonesia']
data_covid.describe()
```
